```python
df_country=pd.read_excel('/content/Country-Code.xlsx')
df_country.head()
```

```python
pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values # Here were are actually getting the list of the actual values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
# Observation=Zomatoes Maximum recors are from India and then from USA and UK
```

```python
rating=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating
```

```python
# Conculation: When rating is between 4.5 to 4.9 --->Excellent
# When rating is between 4.0 to 4.4---> Very Good
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=rating,hue='Rating color')
```

```python
sns.countplot(x='Rating color',data=rating,palette=['blue','red','orange','yellow','green','green'])
# Y-Axis is the frequency
```
